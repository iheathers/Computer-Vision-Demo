#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
another_copy = converted_image.copy()
```

```python
plt.imshow(another_copy)
```

```python
import numpy as np
points = np.array([
    [400,350],
    [250, 700],
    [600, 700]
])

adjusted_points = points.reshape(-1, 1, 2)

cv2.polylines(another_copy, [adjusted_points], isClosed=True, color=(0, 0, 255), thickness=5)
```

### BONUS TASK. Can you figure our how to fill in this triangle? It requires a different function that we didn't show in the lecture! See if you can use google search to find it.

[CLICK ME FOR A DIRECT LINK TO THE HINT](https://docs.opencv.org/3.0-beta/modules/imgproc/doc/drawing_functions.html#fillpoly)

```python
import numpy as np
points = np.array([
    [400,350],
    [250, 700],
    [600, 700]
])

adjusted_points = points.reshape(-1, 1, 2)

cv2.fillPoly(another_copy, [adjusted_points], color=(0, 0, 255))
```

#### TASK: (NOTE: YOU WILL NEED TO RUN THIS AS A SCRIPT). Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.

```python
import cv2
import numpy as np


# Callback function 

def draw_circle(event, x, y, flags, param):
    
    if event == cv2.EVENT_LBUTTONDOWN:        
        print("Left Button Clicked")        
        cv2.circle(image, center=(x,y), radius=20, thickness=2, color=(255, 0, 0))
        
    if event == cv2.EVENT_RBUTTONDOWN:
        print("Right button Clicked")
        cv2.circle(image, center=(x,y), radius = 20, thickness = 2, color=(0, 0, 255))
        
    
cv2.namedWindow('Drawing_image_with_mouse')
cv2.setMouseCallback('Drawing_image_with_mouse', draw_circle)

# Display image

image = cv2.imread('DATA/dog_backpack.jpg')

while True:
    
    cv2.imshow("Drawing_image_with_mouse", image )
    
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
