```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("DATA/00-puppy.jpg")
```

```python

plt.style.use("dark_background")
fig, ax = plt.subplots(figsize=(10, 7))
ax.imshow(img)
```

```python
fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python

fig, ax = plt.subplots(figsize=(10, 7))
ax.imshow(fixed_img)
```

```python
img_gray = cv2.imread('DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python

fig, ax = plt.subplots(figsize=(10,15))
ax.imshow(img_gray, cmap="gray")
```

```python
resized_img = cv2.resize(fixed_img, dsize=(100, 100))
```

```python
plt.imshow(resized_img)
```

```python
w_ratio = 0.5
h_ratio =  0.5
```

```python
resized_img_ratio = cv2.resize(fixed_img, (0, 0), fixed_img, w_ratio, h_ratio)
```

```python
plt.imshow(resized_img_ratio)
```
