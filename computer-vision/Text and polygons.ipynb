```python
blank_canvas = np.zeros(shape=(1024, 1024, 3))
```

```python

fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(blank_canvas)

points = np.array([[100, 100], [200, 400] , [300, 300], [500, 200]], np.int32)

points = points.reshape(-1, 1, 2)


cv2.polylines(blank_canvas, [points], color=(0, 122, 223), thickness=3, isClosed=True)
ax.imshow(blank_canvas)
```
