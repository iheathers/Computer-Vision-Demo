#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../Data/dog_backpack.jpg')
```

```python
plt.imshow(img)
```

```python
fixed_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fixed_img)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
plt.imshow(cv2.flip(fixed_img,0))
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
plt.imshow(cv2.rectangle(fixed_img,pt1=(200,380),pt2=(600,700),color=(255,0,0),thickness=10))
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
vertices = np.array([[250,700],[400,400],[600,700]],dtype = np.int32)
pts = np.reshape(vertices,(-1,1,2))
pts.shape
```

```python
o_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cv2.polylines(o_img,[pts],isClosed=True,color=(0,0,255),thickness = 10))
```

```python
o_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cv2.fillPoly(o_img,[pts],color=(0,0,255)))
```

#### TASK: (NOTE: YOU WILL NEED TO RUN THIS AS A SCRIPT). Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.

```python
import numpy as np
import cv2

def draw_circle(event,x,y,flags,param):
    
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,0,255),5)

cv2.namedWindow(winname='task')
cv2.setMouseCallback('task',draw_circle)
while True:
    cv2.imshow('task',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
